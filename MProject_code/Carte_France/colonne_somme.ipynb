```python
import pandas as pd
```

```python
df = pd.read_csv('output.csv')
```

```python
df = df[df.annais != 'XXXX']
```

```python
df = df[df.dpt !='XX']
```

```python
df.head(10)
```

```python
df.shape
```

```python
df = df.groupby(['annais','dpt'])[['nombre']].sum()
```

```python
df.head(100)
```

```python
df = df.merge(df.groupby([df.annais, df.dpt]).sum().rename(columns={'Result':'somme'}), 
... how='left', 
... left_on=[df.annais, df.dpt], 
... right_index=True) # When you do groupby, the columns you group by will be in the index
```

```python
df.head()
```

```python
df = df.drop(['nombre_x','sexe_y'], axis=1)
```

```python
df.columns = ['sexe', 'preusuel', 'annais', 'dpt', 'somme']
```

```python
df.head()
```

```python
df.to_csv('newCSV.csv', sep=',', index = False)
```
